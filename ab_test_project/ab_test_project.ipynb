## Приоритизация гипотез

### Фреймворк ICE для приоритизации гипотез

### Описание метода ICE

ICE (от англ. impact, confidence, effort / ease «влияние, уверенность, усилия / простота») — один из самых популярных способов приоритизации задач:



- Impact — насколько сильно изменение повлияет на пользователей, их опыт и удовлетворение от продукта;


- Confidence — насколько вы уверены, что это изменение повлияет на пользователей именно так;


- Efforts — сколько стоит протестировать эту гипотезу.


Параметры фреймворков ICE и RICE оценивают в удобных шкалах: например, от 0 до 10 или в числах ряда Фибоначчи.

Посчитаем **ICE** (impact, confidence, effort) для гипотез и добавим его в виде нового столбца 'ICE' к существующей таблице. Выведем данные отсортированные по убыванию их приоритета

```python
#создаем новый столбец с подсчитанными данными на основании формулы 
hypothesis['ICE'] = ((
              hypothesis['impact']*
              hypothesis['confidence']
               )/
              hypothesis['efforts'])
hypothesis['ICE'] = hypothesis['ICE'].round(2)
# добавим столбец с рангом гипотезы по рейтингу ICE 
hypothesis['ICE_rank'] = hypothesis['ICE'].rank(ascending=False).astype(int)
pd.options.display.max_colwidth = 130
print(hypothesis[['hypothesis', 'ICE', 'ICE_rank']].sort_values(by ='ICE', ascending = False))
```

### Примените фреймворк RICE для приоритизации гипотез. Отсортируйте их по убыванию приоритета.

### Описание метода RICE 

RICE практически идентичен методу ICE, но в этом случае добавляется еще один параметр:

- Reach — скольких пользователей затронет изменение, которое вы хотите внести;




Посчитаем **RICE** (reach, impact, confidence, effort) для гипотез и добавим его в виде нового столбца 'RICE' к существующей таблице. Выведем данные отсортированные по убыванию их приоритета

```python
hypothesis['RICE'] = ((hypothesis['reach']*
                hypothesis['impact']*
                hypothesis['confidence']
                /hypothesis['efforts']))
# добавим столбец с рангом гипотезы по рейтингу RICE 
hypothesis['RICE'] = hypothesis['RICE'].round(2)
hypothesis['RICE_rank'] = hypothesis['RICE'].rank(ascending=False).astype(int)
pd.options.display.max_colwidth = 130
print(hypothesis[['hypothesis', 'RICE', 'RICE_rank']].sort_values(by = 'RICE', 
                                                                  ascending = False))
```

```python
# считаем по группе А кол во посетителей по дате 
visitorsADaily_filtered = visitors[visitors['group'] == 'A'][['date',
                                                              'visitors']]
# переименовываем столбцы
visitorsADaily_filtered.columns = ['date', 'visitorsPerDateA_filtered']
# суммироем кол-во посетителей по группе А 
visitorsACummulative_filtered = visitorsADaily_filtered.apply(
    lambda x: visitorsADaily_filtered[visitorsADaily_filtered['date'] <= x['date']].agg(
        {'date': 'max', 'visitorsPerDateA_filtered': 'sum'}
    ),
    axis=1,
)
# переименовываем столбцы 
visitorsACummulative_filtered.columns = ['date', 'visitorsCummulativeA_filtered']

# считаем по группе B кол во посетителей по дате 
visitorsBDaily_filtered = visitors[visitors['group'] == 'B'][['date', 
                                                              'visitors']]
# переименовываем столбцы
visitorsBDaily_filtered.columns = ['date', 'visitorsPerDateB_filtered']
# суммироем кол-во посетителей по группе B 
visitorsBCummulative_filtered = visitorsBDaily_filtered.apply(
    lambda x: visitorsBDaily_filtered[visitorsBDaily_filtered['date'] <= x['date']].agg(
        {'date': 'max', 'visitorsPerDateB_filtered': 'sum'}
    ),
    axis=1,
)
# переименовываем столбцы
visitorsBCummulative_filtered.columns = ['date', 'visitorsCummulativeB_filtered']

# суммарная выручка по пользователям А 
ordersADaily_filtered = (
    orders_filtered[orders_filtered['group'] == 'A'][['date', 'transactionId', 'visitorId', 'revenue']]
    .groupby('date', as_index=False)
    .agg({'transactionId': pd.Series.nunique, 'revenue': 'sum'})
)
# переименовываем столбцы
ordersADaily_filtered.columns = ['date', 'ordersPerDateA_filtered', 'revenuePerDateA_filtered']
ordersADaily_filtered.columns = ['date', 'ordersPerDateA_filtered', 'revenuePerDateA_filtered']
# суммарная выручка по пользователям А 
ordersACummulative_filtered = ordersADaily_filtered.apply(
    lambda x: ordersADaily_filtered[ordersADaily_filtered['date'] <= x['date']].agg(
        {'date': 'max', 'ordersPerDateA_filtered': 'sum', 'revenuePerDateA_filtered': 'sum'}
    ),
    axis=1,
).sort_values(by=['date'])
ordersACummulative_filtered.columns = [
    'date',
    'ordersCummulativeA_filtered',
    'revenueCummulativeA_filtered',
]
# суммарная выручка по пользователям B
ordersBDaily_filtered = (
    orders_filtered[orders_filtered['group'] == 'B'][['date', 'transactionId', 'visitorId', 'revenue']]
    .groupby('date', as_index=False)
    .agg({'transactionId': pd.Series.nunique, 'revenue': 'sum'})
)
ordersBDaily_filtered.columns = ['date', 'ordersPerDateB_filtered', 'revenuePerDateB_filtered']
# суммарная выручка по пользователям B
ordersBCummulative_filtered = ordersBDaily_filtered.apply(
    lambda x: ordersBDaily_filtered[ordersBDaily_filtered['date'] <= x['date']].agg(
        {'date': 'max', 'ordersPerDateB_filtered': 'sum', 'revenuePerDateB_filtered': 'sum'}
    ),
    axis=1,
).sort_values(by=['date'])
# переименовываем столбцы
ordersBCummulative_filtered.columns = [
    'date',
    'ordersCummulativeB_filtered',
    'revenueCummulativeB_filtered',
]
# сводная таблица
data_filtered = (
    ordersADaily_filtered.merge(
        ordersBDaily_filtered, left_on='date', right_on='date', how='left'
    )
    .merge(ordersACummulative_filtered, left_on='date', right_on='date', how='left')
    .merge(ordersBCummulative_filtered, left_on='date', right_on='date', how='left')
    .merge(visitorsADaily_filtered, left_on='date', right_on='date', how='left')
    .merge(visitorsBDaily_filtered, left_on='date', right_on='date', how='left')
    .merge(visitorsACummulative_filtered, left_on='date', right_on='date', how='left')
    .merge(visitorsBCummulative_filtered, left_on='date', right_on='date', how='left')
)

#print(data_filtered.head(5)) 
```

### Cтатистическая значимость различий в конверсии между группами по «сырым» данным

### Способы проверки:
- Критерий Манна-Уитни

Создаем сэмплы для проверки данных

```python
ordersByUsersA = (
    orders_filtered[orders_filtered['group'] == 'A']
    .groupby('visitorId', as_index = False)
    .agg({'transactionId' : pd.Series.nunique})
)
ordersByUsersA.columns = ['visitorId', 'orders']

ordersByUsersB = (
    orders_filtered[orders_filtered['group'] == 'B']
    .groupby('visitorId', as_index = False)
    .agg({'transactionId' : pd.Series.nunique})
)
ordersByUsersB.columns = ['visitorId', 'orders']

sampleA = pd.concat([ordersByUsersA['orders'],
                     pd.Series(0,
                               index=np.arange(data_filtered['visitorsPerDateA_filtered']
                                               .sum() - 
                                               len(ordersByUsersA['orders']
                                                  )
                                              ),
                               name='orders')
                    ],axis=0
                   )

sampleB = pd.concat([ordersByUsersB['orders'],
                     pd.Series
                     (0,
                      index=np.arange(data_filtered['visitorsPerDateB_filtered']
                                      .sum()-
                                      len(ordersByUsersB['orders']
                                         )
                                     )
                      , name='orders')
                    ],axis=0
                   )
```

#### - H0 - Статистически значимых различий в конверсии между группами A и B нет;

#### - H1 - Статистически значимые различия в конверсии между группами есть.

#### - Уровень значимости 0.05

```python
alpha = 0.05
print("p_value = {0:.4f}"
      .format(st.mannwhitneyu
              (sampleA, sampleB,
               alternative = 'two-sided')[1]
             )
     )
print("относительный прирост конверсии по группе В: {0:.4f}"
      .format(
          sampleB.mean() 
          / sampleA.mean() 
          - 1)
     )

if st.mannwhitneyu(sampleA,
                   sampleB,
                   alternative = 'two-sided')[1] > alpha:
    print('Статистически значимых различий в конверсии между группами нет, гипотезу не отвергаем')
else:
    print('Отвергаем нулевую гипотезу, есть различия в конверсии между группами')
```

### Вывод по статистической значимости различий в конверсии между группами по «сырым» данным: 

- Отвергаем нулевую гипотезу, есть различия в конверсии между группами


- Относительный прирост конверсии по группе В: более 17% 

### Статистическая значимость различий в среднем чеке заказа между группами по «сырым» данным.

### Способы проверки:
- Критерий Манна-Уитни

#### - H0 - Статистически значимых различий в среднем чеке заказа между группами A и B нет;

#### - H1 -  Статистически значимые различия в среднем чеке заказа между группами A и B есть;

#### - Уровень значимости 0.05

```python
alpha = 0.05
print("p_value = {0:.4f}"
      .format(st.mannwhitneyu
              (orders_filtered[orders_filtered['group'] == 'A']['revenue'],
               orders_filtered[orders_filtered['group'] == 'B']['revenue'],
               alternative = 'two-sided')[1]
             )
     )
print("Относительный прирост среднего чека по группе В: {0:.4f}"
      .format(orders_filtered[orders_filtered['group']=='B']
              ['revenue'].mean()
              /orders_filtered[orders_filtered['group']=='A']['revenue'].mean()-1
             )
     )

if st.mannwhitneyu(orders_filtered[orders_filtered['group'] == 'A']['revenue'],
                   orders_filtered[orders_filtered['group'] == 'B']['revenue'],
                   alternative = 'two-sided')[1] > alpha:
    print('Статистически значимых различий в среднем чеке заказа между группами A и B нет, гипотезу не отвергаем')
else:
    print('Отвергаем нулевую гипотезу, статистически значимые различия в среднем чеке между группами есть')
```

### Статистическая значимость различий в конверсии между группами по «очищенным» данным

#### Чистка данных по заранее обозначеным критериям

####  Критерии аномальности: 
- кол-во заказов не больше 2
- сумма заказа не выше 53904.

```python
abnormal_orders = pd.concat(
    [ordersByUsersA[ordersByUsersA['orders'] >= np.percentile(orders_visitor_filtered['orders'],[95, 99])[1]]['visitorId'],
    ordersByUsersB[ordersByUsersB['orders'] >= np.percentile(orders_visitor_filtered['orders'],[95, 99])[1]]['visitorId']
    ],
    axis = 0
)

abnormal_revenue = orders_filtered[orders_filtered['revenue'] > np.percentile(orders_filtered['revenue'],[99])[0]]['visitorId']
abnormal_users = (
    pd.concat([abnormal_orders, 
               abnormal_revenue],
              axis=0
             ).drop_duplicates()
    .sort_values()
)
# создаем сэмплы
sample_A_cleared = pd.concat([ordersByUsersA
                               [np.logical_not(ordersByUsersA['visitorId']
                                               .isin(abnormal_users
                                                    )
                                              )
                               ]['orders']
                               , pd.Series
                               (0,
                                index=np.arange(data_filtered['visitorsPerDateA_filtered']
                                                .sum() - 
                                                len(ordersByUsersA['orders']
                                                   )
                                               ), 
                                name='orders')]
                              , axis=0
                             )

sample_B_cleared = pd.concat([ordersByUsersB
                               [np.logical_not(ordersByUsersB['visitorId']
                                               .isin(abnormal_users
                                                    )
                                              )
                               ]['orders']
                               , pd.Series
                               (0, 
                                index=np.arange(data_filtered['visitorsPerDateB_filtered']
                                                .sum() - 
                                                len(ordersByUsersB['orders']
                                                   )
                                               ),
                                name='orders')]
                              , axis=0
                             )

```

### Способы проверки:
- Критерий Манна-Уитни

#### - H0 - Статистически значимых различий в конверсии между группами A и B нет;


#### - H1 - Статистически значимые различия в конверсии между группами есть.


#### - Уровень значимости 0.05

```python
alpha = 0.05
print("p_value = {0:.4f}"
      .format(st.mannwhitneyu
              (sample_A_cleared, sample_B_cleared,
               alternative = 'two-sided')[1]
             )
     )
print("относительный прирост конверсии по группе В: {0:.4f}"
      .format(
          sampleB.mean() 
          / sample_A_cleared.mean() 
          - 1)
     )

if st.mannwhitneyu(sample_A_cleared,
                   sample_B_cleared,
                   alternative = 'two-sided')[1] > alpha:
    print('Статистически значимых различий в конверсии между группами нет, гипотезу не отвергаем')
else:
    print('Отвергаем нулевую гипотезу, есть различия в конверсии между группами')
```

### Вывод по статистической значимости различий в конверсии между группами по «очищенным» данным: 

- Отвергаем нулевую гипотезу, есть различия в конверсии между группами

### Статистическая значимость различий в среднем чеке заказа между группами по «очищенным» данным

### Способы проверки:

#### - H0 - Статистически значимых различий в среднем чеке заказа между группами A и B нет;

#### - H1 -  Статистически значимые различия в среднем чеке заказа между группами A и B есть;

#### - Уровень значимости 0.05

```python
alpha = 0.05
# данные по среднему чеку
mean_check_A = orders_filtered[np.logical_and(orders_filtered['group'] == 'A',
                                              np.logical_not(orders_filtered['visitorId']
                                                             .isin(abnormal_users)
                                                            )
                                             )]['revenue']
mean_check_B = orders_filtered[np.logical_and(orders_filtered['group'] == 'B',
                                              np.logical_not(orders_filtered['visitorId']
                                                             .isin(abnormal_users)
                                                            )
                                             )]['revenue']

# применение критерия для определения различий в среднем чеке по очищенным данным 
print("p_value = {0:.4f}".format(st.mannwhitneyu(mean_check_A,
                                                 mean_check_B,
                                                 alternative = 'two-sided')[1])
     )
print("Относительный прирост среднего чека по группе В: {0:.4f}".format(mean_check_B.mean()
                                                                      /mean_check_A.mean()
                                                                      -1)
     )

if st.mannwhitneyu(mean_check_A, mean_check_B, alternative = 'two-sided')[1] > alpha:
    print('Статистически значимых различий в среднем чеке заказа между группами A и B нет, гипотезу не отвергаем')
else:
    print('Отвергаем нулевую гипотезу, статистически значимые различия в среднем чеке между группами есть')
```

- Статистически значимых различий в среднем чеке заказа между группами A и B нет, гипотезу не отвергаем


- Относительный прирост конверсии по группе В: 0.0024


- Аномально дорогие заказы очень сильно повлияли на средний чек группы B. В их отсутствии он ниже чем у группы A
