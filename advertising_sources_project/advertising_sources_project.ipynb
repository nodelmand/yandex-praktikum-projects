- по способу достижения первого уровня (проект/победа над врагом/без способа)

#### По способу достижения первого уровня:

```python
# Создаем список с id окончавших проект
projec_list = (game_actions
               .query('event == "project"')['user_id']
               .tolist()
              )
# Создаем список с id победителей врагов
winner_list = (game_actions
               .loc[
                   (game_actions['event']
                    == "finished_stage_1"
                   )
                              ]
               ['user_id']
               .reset_index()
               .query('user_id != @projec_list')
              )
winner_list = (winner_list['user_id']
               .tolist()
              )
# создаем функцию, которая фильтрует
# и присваивает игрокам категорию
def categorizer(user_id):
    if user_id in projec_list:
        return 'project'
    if user_id in winner_list:
        return 'win'
    return 'not_finished'
# Создаем столбец с категорией 
game_actions['player_category'] = game_actions['user_id'].apply(categorizer)
```

#### По скорости достижения первого уровня

Разделим игроков на след категории:

- fast_fin - те, кто уложились в первый квартиль (q1, 25%) переменной period


- slow_fin - закончили уровень дольше чем первый квартиль (q1) переменной period (те кто быстро завершлили)


- not_fin - не закончили первый уровень за время внутри таблицы 


```python
# рассмотрим сколько времени тратят игроки от первого действия до окончания первого уровня
start_date = (game_actions
                 .groupby('user_id')
                 .agg({'event_datetime':'min'})
                 .reset_index()
                )
# датасет с первым действием игрока
finish_date = (game_actions
                  .groupby('user_id')
                  .agg({'event_datetime':'max'})
                  .reset_index()
                 ) 
#датасет 
#с последним действием игрока
df = start_date.merge(finish_date,
                              on = 'user_id'
                             )
#объединяем таблицы 
 #переименовываем название столбцов
df.rename(columns = {'event_datetime_x': 'start_datetime', 
                     'event_datetime_y': 'finish_datetime'},
          inplace = True
         )
#убираем из даты время
df['start_datetime'] = (df['start_datetime']
                        .astype('datetime64[D]')
                       )
#убираем из даты время
df['finish_datetime'] = (df['finish_datetime']
                         .astype('datetime64[D]')
                        )
#вычисляем период
df['period'] = (df['finish_datetime']
                - df['start_datetime']
               ) #вычисляем период
#
df['period'] = df['period'].dt.days
#
df['period'] = df['period'].astype('int')
##объединяем таблицы
df = df.merge(game_actions[{'event_datetime',
                            'event',
                            'building_type', 
                            'user_id', 
                            'project_type',
                            'player_category',
                            }
                          ], 
              on = 'user_id', 
              how = 'inner'
             )
#
df = df.drop_duplicates() 
#
df['tempo'] = 'fast_fin' 
# создаем временный df
# в нем методом describe записываем в переменную q1 перцентиль 25% 
# для категории 'project'
desctibe_df = df.groupby('player_category').agg({'period':'describe'}) 
desctibe_df.columns = ['count',
                      'mean',
                      'std',
                      'min',
                      '25',
                      '50',
                      '75',
                      'max']
# 
q1 = desctibe_df["25"][1]
df.loc[df['period'] > q1,
            'tempo'] = 'slow_fin'
df.loc[df['player_category'] == 'not_finished','tempo'] = 'not_fin'
```

### WAU/DAU 
Для рассчета недельных и дневных показателей количества игроков и их действий создаем столбцы внутри таблицы с выделенными неделей и днем. По этим столбцам будем считать метрики WAU и DAU. Информация по метрикам *(https://blog.popcornmetrics.com/5-user-engagement-metrics-for-growth/)

#### Количество активных игроков в неделю и в день

```python
# Устанавливаем значения для палетки
sns.set_palette("Paired")
# выделяем столбец с днем
df['event_dt'] = pd.to_datetime(df['event_datetime']).dt.normalize()
#  выделяем столбец с неделей
df['event_week'] = df['event_datetime'].dt.week
# кол-во активных игроков в неделю
wau_total = (
    df.groupby('event_week')
    .agg({'user_id': 'nunique'})
    .mean()
)
# кол-во активных игроков в день
dau_total = (
    df.groupby('event_dt')
    .agg({'user_id': 'nunique'})
    .mean()
)
# создаем словрь 
dict1 = ({'event_dt': 'days', 
          'event_week': 'weeks'}
        )
label1 = ['day',
          'week'
         ]

plt.figure(figsize = (15, 2))
a = 1
b = 0
for key in dict1:
    (
    df.groupby(key)
        .agg({'user_id': 'nunique'})
        .plot(grid = True, 
              ax = plt.subplot(1,
                               2, 
                               a)
             )
    )  
    plt.legend()
    plt.title('Динамика количества пользователей в игре в {}'.format(label1[b]),
              fontsize = 12
              , fontweight='bold'
             )
    plt.xlabel('{}'.format(dict1[key]),
               fontsize = 11
              ) 
    plt.ylabel('Количество пользователей',
               fontsize = 11
               , fontweight='bold'
              )
    a += 1
    b += 1
plt.show()

print('Среднее кол-во игроков в день:', int(dau_total))
print('Среднее кол-во игроков в неделю:', int(wau_total))
```

### Расчет CAC (customer acquisition cost)

```python
# выделяем дату 
df['first_date'] = df['start_datetime']
# дата привлечения пользователя всегда больше на 1
ad_costs['first_date'] = ad_costs['day'] + timedelta(days=1)

# игроки по дням и источникам
source_df = df.groupby(['first_date',
                                'source'])['user_id'].nunique().reset_index()

# создаем df с костами
costs = ad_costs.merge(source_df, on = ['first_date',
                                         'source']).rename(columns = {'user_id': 'users'})

# расчет cac
costs['cac'] = (costs['cost'] / 
                costs['users']
               ).round(2)

# добавление столбца с CAC в таблицу df
df = (
    df.merge(costs, on = ['first_date',
                          'source'],
               how = 'left'
              )
        .drop(['day',
               'users',
               'cost'],
              axis = 1
             )        
)
# создадим профили пользователей
players_cac = df[['user_id', 
                 'first_date',
                 'source']
               ].drop_duplicates()

# добавление CAC в профили пользователей
players_cac = (
    players_cac.merge(costs, on = ['first_date',
                              'source'], 
                   how = 'left'
                  )
        .drop(['day', 
               'users',
               'cost'],
              axis = 1
             )        
)
```

```python
pivot = (
    players_cac.groupby('source')['cac']
        .mean()
        .round(2)
        .sort_values(ascending=False)
        .to_frame(name='mean_amount').reset_index()
)
sns.barplot(x = 'source', 
            y = 'mean_amount',
            data = pivot
           )
plt.xticks(rotation = 45)
plt.xlabel('Source'
           , fontweight='bold')
plt.ylabel('Средняя цена за игрока'
          , fontweight='bold')
plt.title('Средний CAC по каналам'
         , fontweight='bold')
plt.show()
pivot
```

# Проверка гипотез

##  Гипотеза: "Время завершения уровня различается в зависимости способа прохождения"

#### Нулевая гипотеза H0 :
- Время завершения уровня в зависимости от способа прохождения одинаковое

#### Альтернативная двухстороняя гипотеза H1 :
- Время завершения уровня в зависимости от способа прохождения различаются

Критический уровень значимости - 0.05

```python
var1 = df[df['player_category'] == 'win']['period']
var2 = df[df['player_category'] == "project"]['period']

alpha = 0.05
results = st.ttest_ind(var1,var2)
print('Среднее ожидаемое значения:', results)
print('p-значение:', results.pvalue)
if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергуть нулевую гипотезу')
```

##  Гипотеза: "Клиенты, привлеченные по каналу yandex_direct  переходят на первый уровень быстрее, чем по каналу instagram_new_adverts "

#### Нулевая гипотеза H0 :
- Клиенты переходят одинаково быстро
#### Альтернативная двухстороняя гипотеза H1 :
- Клиенты переходят с разной скоростью

Критический уровень значимости - 0.05

```python
var1 = df[df['source'] == 'yandex_direct']['period']
var2 = df[df['source'] == "instagram_new_adverts"]['period']

alpha = 0.5
results = st.ttest_ind(var1,var2)
print('Среднее ожидаемое значения:', results)
print('p-значение:', results.pvalue)
if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергуть нулевую гипотезу')
```
